```python
import pandas as pd
import numpy as np
```

```python
#Dictionary of lists
dir = {"name":["abc" , "xyz" , "pqr"],
       "age":[23,34,45],
       "city":["Bikaner" , "Ahemdabad" , "jaipur"],
       "marks":[22,33,44]
       }

df =pd.DataFrame(dir, index=["a","b" ,"c"])
print(df)
```

```python
#Lists of Lists
L = [["abc" , "xyz" , "pqr"] ,[23,34,45] ,["Bikaner" , "Ahemdabad" , "jaipur"], [22,33,44]]
df = pd.DataFrame(L , index=["name" , "age" , "city" , "marks"] , columns=["a","b" ,"c"])
print(df)
```

```python
#list of dictionaries
L = [{"name":"abc" , "age":23 , "city":"Bikaner" , "marks":22} , {"name":"xyz" , "age":34 , "city":"Ahemdabad" , "marks":33} , {"name":"pqr" , "age":45 , "city":"jaipur" , "marks":44}]
df = pd.DataFrame(L)
print(df)
```

```python
#dictionary of series
dir = {"name":pd.Series(["abc" , "xyz" , "pqr"]),
       "age":pd.Series([23,34,45]),
       "city":pd.Series(["Bikaner" , "Ahemdabad" , "jaipur"]),
       "marks":pd.Series([22,33,44])}
df = pd.DataFrame(dir)
print(df)
```

```python
#numpy array
arr = np.array([["abc" , "xyz" , "pqr"] ,[23,34,45] ,["Bikaner" , "Ahemdabad" , "jaipur"], [22,33,44]])
df = pd.DataFrame(arr)
print(df)
```

```python
#Empty values in dataframe
df = pd.DataFrame()
print(df)
```

```python
# set column as index
df = pd.DataFrame({
    "id": [11,22,33],
    "Name": ["kartik" , "abc" , "xyz"],
    "Marks":[34,67,89]
})
print(df)
print(df.set_index(keys="id")) # it's just showing the output but not changing the dataframe


# mordifying the dataframe
df2 = df.set_index(keys="id") # now the dataframe is modified and id is set as index
print(df2)

# mordifing the orignal dataframe
df.set_index(keys="id" , inplace=True) # now the original dataframe is modified and id is set as index
print(df)


```

```python
df = pd.DataFrame({
    "id": [11,22,33],
    "Name": ["kartik" , "abc" , "xyz"],
    "Marks":[34,67,89]
})
print(df)
# set column as index without droping 
df2 = df.set_index(keys="id" , drop=False) # now the dataframe is modified and id is set as index but the id column is not dropped
print(df2)
```

DataFrame Attribute

```python

df = pd.DataFrame({
    "id": [11,22,33 ,34],
    "Name": ["kartik" , "abc" , "xyz" ,"hhjsd"],
    "Marks":[34,67,89 ,54]
})
print(df)
# Shape
print(df.shape) # it will give us the number of rows and columns in the dataframe as a tuple (4,3) means 4 rows and 3 columns

#Size
print(df.size) # it will give us the total number of elements in the dataframe as a single number 12 means 4 rows and 3 columns = 12 elements

#ndim
print(df.ndim) # it will give us the number of dimensions in the dataframe as a single number 2 means 2 dimensions (rows and columns)

#columns
print(df.columns) # it will give us the column names of the dataframe as an index object Index(['id', 'Name', 'Marks'], dtype='object')

#index
print(df.index) # it will give us the index of the dataframe as an index object RangeIndex(start=0, stop=4, step=1)

#dtypes
print(df.dtypes) # it will give us the data types of each column in the dataframe as a series id        int64
              # Name     object
              # Marks    int64
              # dtype: object

#values
print(df.values)
```
