DataFrame Attribute

```python

df = pd.DataFrame({
    "id": [11,22,33 ,34],
    "Name": ["kartik" , "abc" , "xyz" ,"hhjsd"],
    "Marks":[34,67,89 ,54]
})
print(df)
print("-------------------------------------------------------------------------------")



# Shape
print(df.shape) # it will give us the number of rows and columns in the dataframe as a tuple (4,3) means 4 rows and 3 columns
print("-------------------------------------------------------------------------------")



#Size
print(df.size) # it will give us the total number of elements in the dataframe as a single number 12 means 4 rows and 3 columns = 12 elements
print("-------------------------------------------------------------------------------")



#ndim
print(df.ndim) # it will give us the number of dimensions in the dataframe as a single number 2 means 2 dimensions (rows and columns)
print("-------------------------------------------------------------------------------")



#columns
print(df.columns) # it will give us the column names of the dataframe as an index object Index(['id', 'Name', 'Marks'], dtype='object')
print("-------------------------------------------------------------------------------")



#index
print(df.index) # it will give us the index of the dataframe as an index object RangeIndex(start=0, stop=4, step=1)
print("-------------------------------------------------------------------------------")



#dtypes
print(df.dtypes) # it will give us the data types of each column in the dataframe as a series id        int64
              # Name     object
              # Marks    int64
              # dtype: object
print("-------------------------------------------------------------------------------")



#values
print(df.values)
print("-------------------------------------------------------------------------------")



#transpose
print(df.T) # it will give us the transpose of the dataframe means rows will become columns and columns will become rows
print("-------------------------------------------------------------------------------")



#Empty
print(df.empty) # it will give us a boolean value True if the dataframe is empty and False if the dataframe is not empty
print("-------------------------------------------------------------------------------")



#axes
print(df.axes)# it will give us the list of index and columns in the dataframe as a list [RangeIndex(start=0, stop=4, step=1), Index(['id', 'Name', 'Marks'], dtype='object')]
print("-------------------------------------------------------------------------------")


#memory usage
print(df.memory_usage()) # it will give us the memory usage of each column in the dataframe
print("-------------------------------------------------------------------------------")

#info
print(df.info()) # it will give us the summary of the dataframe including the number of non-null values in each column and the data types of each column
print("-------------------------------------------------------------------------------")
```

```python
#head
import pandas as pd
import numpy as np
df = pd.DataFrame({
 "ID": [1, 2, 2, 3, 4, 4],
 "Name": ["Amit", "Rahul", "Rahul", "Neha", "Priya", "Priya"],
 "Marks": [78, 85, 85, np.nan, 90, 90],
 "City": ["Ahmedabad", "Surat", "Surat", "Baroda", "Rajkot", "Rajkot"]
})
print(df.head(2)) # it will give us the first 2 rows of the dataframe

#tail
print(df.tail(3)) # it will give us the last 3 rows of the dataframe
#sample
print(df.sample(4)) # it will give us 4 random rows from the dataframe

print(df.sample(frac=False))
#print(df.sample(frac=False))
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
 "ID": [1, 2, 2, 3, 4, 4],
 "Name": ["Amit", "Rahul", "Rahul", "Neha", "Priya", "Priya"],
 "Marks": [78, 85, 85, np.nan, 90, 90],
 "City": ["Ahmedabad", "Surat", "Surat", "Baroda", "Rajkot", "Rajkot"]
})

#info
df.info()


#describe
print("-"*50)
print(df.describe())

print("-"*50)
print(df.describe(percentiles=[0.2,0.3]))

print("-"*50)
print(df.describe(include="all"))

print("-"*50)
print(df.describe(exclude="int64"))

print("-"*50)
print(df.describe(exclude=["int64","float64"]))
```

UNIQUE & FREQUENCY METHODS

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
 "ID": [1, 2, 2, 3, 4, 4],
 "Name": ["Amit", "Rahul", "Rahul", "Neha", "Priya", "Priya"],
 "Marks": [78, 85, 85, np.nan, 90, 90],
 "City": ["Ahmedabad", "Surat", "Surat", "Baroda", "Rajkot", "Rajkot"]
})

#unique() 
print(df["City"].unique()) # it will give us the unique values in the City column as an array ['Ahmedabad' 'Surat' 'Baroda' 'Rajkot']
print("-"*50)

#nunique()
print(df["City"].nunique()) # it will give us the number of unique values in
print("-"*50)

#value_counts()
print(df["City"].value_counts()) # it will give us the count of each unique value in the City column as a series
print("-"*50)

print(df["City"].value_counts(normalize=True)) # it will give us the percentage of each unique value in the City column as a series
print("-"*50)

print(df.value_counts(ascending=False)) #

```

MISSING DATA HANDLING


```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
 "ID": [1, 2, 2, 3, 4, 4],
 "Name": ["Amit", "Rahul", "Rahul", "Neha", "Priya", "Priya"],
 "Marks": [78, 85, 85, np.nan, np.nan, 90],
 "City": ["Ahmedabad", "Surat", "Surat", "Baroda", "Rajkot", "Rajkot"]
})

#isnull

print(df.isnull())
print(df.isnull().sum())

#fillna
import pandas as pd
import numpy as np
df = pd.DataFrame({
 "Maths": [85, 90, np.nan, 88, np.nan],
 "Science": [78, np.nan, 82, 80, 79]
})
print(df)
df.fillna(df.mean(), inplace=True)# it will fill the missing values with the mean of each column
print(df)

df = pd.DataFrame({
 "Age": [20, None, 22],
 "City": ["Ahmedabad", None, "Surat"]
})
df.fillna({"Age": 21, "City": "Unknown"}, inplace=True)
print(df)
```
